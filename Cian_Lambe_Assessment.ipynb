```python
df=pd.read_csv('Accenture_stock_history.csv')
```

```python
import datetime
```

```python
# convert date format
```

```python
df['Date']=df['Date'].apply(lambda x:datetime.datetime.strptime(x,'%d/%m/%Y'))
```

```python
df.head()
```

```python
#remove index
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
#show the movement of the closing share price
```

```python
df['Close'].plot()
plt.show
```

```python
#compare the opening versus closing
```

```python
df['Open'].plot(label = 'ACN Open',figsize=(15,7))
df['Close'].plot(label = 'ACN Close')
plt.legend()
plt.show()
```

```python
#compare the highest versus the lowest
```

```python
df['High'].plot(label = 'ACN High',figsize=(17,5))
df['Low'].plot(label = 'ACN Low')
plt.legend()
plt.title('ACN Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
#calculate represent total money traded
```

```python
df['Total Traded'] = df['Open'] * df['Volume']
```

```python
df.head()
```

```python
df['Total Traded'].plot(label = 'ACN High',figsize=(17,5))
plt.legend()
plt.title('ACN Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
#look at the moving average over a timeperiod to avoid sharp spikes/drops
```

```python
df['Close'].plot(figsize=(15,7))
df['MA50'] = df['Close'].rolling(50).mean()
df['MA50'].plot(label='MA50')
df['MA300'] = df['Close'].rolling(300).mean()
df['MA300'].plot(label='MA300')
plt.legend()
```

```python
from pandas.plotting import scatter_matrix
```

```python
compare = pd.concat([df['Close']],axis =1)
```

```python
scatter_matrix(compare,figsize=(8,8))
```

```python
#Daily percentage change of the stock prices
```

```python
df['returns'] = (df['Close']/df['Close'].shift(1))-1
```

```python
df.head()
```

```python
df['returns'].hist(bins=50,figsize=(15,8))
```

```python
# add in a cumulative return 
```

```python
df['Cumulative Return'] = (1+df['returns']).cumprod()
```

```python
df.head()
```

```python
df['Cumulative Return'].plot(label='ACN', figsize=(15,7))
```
